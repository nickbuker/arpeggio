```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("../data/sd254_users.csv")
```

```python
df.head(10)
```

```python
plt.hist(df["Current Age"])
plt.title("Histogram of Current Age");
```

```python
plt.hist(df["Retirement Age"])
plt.title("Histogram of Retirement Age");
```

```python
df["Gender"].value_counts(dropna=False)
```

```python
df["income"] = df["Yearly Income - Person"].str.replace("$", "").astype("float")
```

```python
plt.hist(df["income"])
plt.title("Histogram of Income");
```

```python
df["debt"] = df["Total Debt"].str.replace("$", "").astype("float")
```

```python
plt.hist(df["debt"])
plt.title("Histogram of Debt");
```

```python
plt.hist(df["FICO Score"])
plt.title("Histogram of FICO Score");
```

```python
df["age_quartile"] = pd.qcut(df["Current Age"], 4)
df["income_quartile"] = pd.qcut(df["income"], 4)
df["debt_quartile"] = pd.qcut(df["debt"], 4)
df["fico_quartile"] = pd.qcut(df["FICO Score"], 4)
```
